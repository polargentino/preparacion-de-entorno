

# 🚀 Guía de Inicio: Entorno de Desarrollo Web

Esta guía te ayudará a configurar las herramientas básicas para comenzar a programar: **Visual Studio Code**, **Git** y **GitHub**. Sigue cada paso con atención.

---

## 1. Instalación y Configuración de Visual Studio Code (VS Code)

Visual Studio Code es el editor de código más utilizado en la industria.

1. **Descarga:** Ve a [code.visualstudio.com](https://code.visualstudio.com/) y descarga la versión para tu sistema operativo.
2. **Instalación:** Ejecuta el instalador. **Importante:** Durante la instalación en Windows, marca las casillas que dicen *"Agregar la acción 'Abrir con Code'"*.
3. **Configurar en Español:**
* Abre VS Code.
* Presiona `Ctrl + Shift + X` (Extensiones).
* Busca **"Spanish Language Pack"** e instálalo.
* Presiona `Ctrl + Shift + P`, escribe `Configure Display Language`, selecciona `es` y reinicia el programa.


4. **Instalar Live Server:**
* En el panel de extensiones (`Ctrl + Shift + X`), busca **"Live Server"** (de Ritwick Dey).
* Haz clic en **Install**. Esto nos permitirá ver los cambios en tiempo real.



---

## 2. Creación del Primer Proyecto

1. **Carpeta de trabajo:** Ve a tu **Escritorio**, haz clic derecho y crea una carpeta nueva llamada `mi-primer-proyecto`.
2. **Abrir en VS Code:**
* Vuelve a VS Code.
* Ve a **Archivo > Abrir carpeta** y selecciona la carpeta que creaste en el escritorio.


3. **Crear el archivo HTML:**
* En el panel de la izquierda (Explorador), haz clic en el icono de **Nuevo archivo**.
* Nómbralo exactamente como `index.html`.
* Escribe `!` y presiona `Tab` o `Enter` para generar la estructura básica de HTML5.


4. **Prueba de Live Server:**
* Escribe un `<h1>Hola Mundo</h1>` dentro del `<body>`.
* Haz clic derecho sobre el código y selecciona **"Open with Live Server"**.
* **Comprobación:** Se debería abrir tu navegador mostrando el mensaje.



---

## 3. Instalación de Git y Uso de la Terminal

Git es el sistema que controla las versiones de nuestro código.

1. **Instalación:** Descárgalo de [git-scm.com](https://git-scm.com/) e instálalo con las opciones por defecto.
2. **Configuración de Terminal en VS Code:**
* Abre la terminal integrada en VS Code con `Ctrl + ñ` (o Terminal > Nueva terminal).
* Asegúrate de que diga **bash** en el desplegable de la derecha.


3. **Comandos Básicos de Bash (Práctica):**
* `pwd`: Muestra en qué carpeta estás posicionado.
* `ls`: Lista los archivos de la carpeta actual (deberías ver `index.html`).
* `mkdir assets`: Crea una carpeta llamada "assets".
* `clear`: Limpia la pantalla de la terminal.



---

## 4. GitHub y Publicación del Proyecto

### Paso A: Crear cuenta y repositorio

1. Crea una cuenta en [github.com](https://github.com/) y verifica tu correo.
2. Haz clic en el botón **"New"** (Nuevo repositorio).
3. Nombre del repositorio: `mi-proyecto-web`.
4. Déjalo como **Public** y no marques ninguna casilla adicional. Haz clic en **Create repository**.

### Paso B: Subir el código (Comandos de Git)

Regresa a la terminal de VS Code y escribe estos comandos uno por uno:

1. **Inicializar:** `git init`
2. **Configurar identidad** (Solo la primera vez):
* `git config --global user.email "tu@email.com"`
* `git config --global user.name "Tu Nombre"`


3. **Preparar archivos:** `git add .`
4. **Primer "guardado":** `git commit -m "Mi primer commit"`
5. **Vincular a GitHub:** Copia la línea que dice `git remote add origin...` de la página de GitHub y pégala en la terminal.
6. **Subir código:** `git push -u origin master` (o `main` según indique GitHub).

### Paso C: Publicar en la Web (GitHub Pages)

1. En tu repositorio de GitHub, ve a **Settings** (Configuración).
2. En el menú lateral, busca **Pages**.
3. En "Branch", selecciona **main** (o master) y la carpeta **/(root)**. Haz clic en **Save**.
4. **Comprobación:** En unos minutos, GitHub te dará un enlace (URL) donde tu `index.html` será visible para todo el mundo.

## 🛠️ Actividad Práctica: "Misión Terminal"

El objetivo es manipular archivos y carpetas sin usar el ratón, solo la terminal de VS Code.

### Reto 1: Organización de archivos

Escribe los siguientes comandos en orden y observa cómo cambia tu panel de archivos a la izquierda:

1. **Verificar ubicación:** `pwd`
*(Asegúrate de que la ruta termine en tu carpeta de proyecto).*
2. **Crear estructura de carpetas:** `mkdir css js img`
*(Esto creará tres carpetas: una para estilos, otra para scripts y una para imágenes).*
3. **Crear archivos desde la terminal:** `touch css/style.css js/script.js`
*(Creamos los archivos directamente dentro de sus carpetas).*
4. **Listar para verificar:** `ls -R`
*(Verás toda tu nueva estructura en pantalla).*

---

### Reto 2: Flujo de Trabajo Real (Git)

Ahora que hemos creado carpetas nuevas, vamos a actualizar tu repositorio en GitHub:

1. **Ver el estado:** `git status`
*(Verás las nuevas carpetas en rojo porque Git aún no las sigue).*
2. **Preparar cambios:** `git add .`
3. **Confirmar cambios:** `git commit -m "Estructura de carpetas inicial creada"`
4. **Sincronizar con la nube:** `git push`

## 🎨 Guía: Separando la Estructura del Diseño

### 1. Crear el archivo CSS

En VS Code, dentro de la carpeta `css` que creamos antes por terminal (o creando una nueva), crea un archivo llamado `style.css`.

### 2. Vincular el CSS con el HTML

Para que el navegador sepa que debe aplicar los estilos, debemos "conectar" los archivos.

**Acción:** Dentro de tu `index.html`, busca la etiqueta `<head>` y, justo antes de que se cierre (`</head>`), pega esta línea:

```html
<link rel="stylesheet" href="css/style.css">

```

> **Nota para los alumnos:** Asegúrate de que la ruta `css/style.css` sea correcta según dónde guardaste el archivo.

---

### 3. Escribir el Código CSS

Abre tu archivo `style.css` y pega el siguiente código. He quitado los estilos que estaban dentro de las etiquetas HTML para que ahora todo se controle desde aquí:

```css
/* 1. Estilos Globales */
body {
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    line-height: 1.6;
    background-color: #f0f2f5;
    color: #333;
    margin: 0;
    padding: 20px;
}

/* 2. Encabezado y Navegación */
header {
    background-color: #2c3e50;
    color: white;
    padding: 2rem;
    text-align: center;
    border-radius: 8px 8px 0 0;
}

nav ul {
    background: #34495e;
    padding: 10px;
    list-style: none;
    display: flex;
    justify-content: center;
    gap: 20px;
}

nav a {
    color: white;
    text-decoration: none;
    font-weight: bold;
}

nav a:hover {
    color: #3498db;
}

/* 3. Secciones */
section {
    background: white;
    padding: 25px;
    margin-top: 20px;
    border-radius: 8px;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1);
}

h2 {
    color: #2c3e50;
    border-bottom: 2px solid #3498db;
    padding-bottom: 10px;
}

/* 4. Tablas */
table {
    width: 100%;
    border-collapse: collapse;
    margin: 20px 0;
}

th, td {
    border: 1px solid #ddd;
    padding: 12px;
    text-align: left;
}

th {
    background-color: #3498db;
    color: white;
}

/* 5. Formularios */
input, textarea {
    width: 100%;
    padding: 10px;
    margin-top: 5px;
    border: 1px solid #ccc;
    border-radius: 4px;
    box-sizing: border-box; /* Muy importante para el ancho */
}

button {
    background-color: #27ae60;
    color: white;
    border: none;
    padding: 12px 20px;
    border-radius: 4px;
    cursor: pointer;
    font-size: 16px;
    transition: background 0.3s;
}

button:hover {
    background-color: #219150;
}

```

```css
/* Estilos base para que se vea limpio */
body {
    font-family: Arial, sans-serif;
    background-color: #e9ecef;
    margin: 0;
    padding: 20px;
    color: #333;
}

header {
    text-align: center;
    margin-bottom: 30px;
}

section {
    background: white;
    padding: 20px;
    margin-bottom: 40px;
    border-radius: 10px;
    box-shadow: 0 4px 10px rgba(0,0,0,0.1);
}

/* Estilos de las "Cajas" del laboratorio */
.contenedor {
    background-color: #dee2e6;
    border: 2px dashed #adb5bd;
    padding: 10px;
    border-radius: 5px;
}

.caja {
    width: 80px;
    height: 80px;
    color: white;
    display: flex;
    justify-content: center;
    align-items: center;
    font-weight: bold;
    font-size: 1.2rem;
    border-radius: 5px;
    background-color: #6c757d; /* Color por defecto */
}

.roja { background-color: #e74c3c; }
.azul { background-color: #3498db; }
.verde { background-color: #2ecc71; }

/* --- AQUÍ EMPIEZA LA MAGIA DE FLEXBOX --- */

/* Ejercicio 1: Cambia justify-content por: flex-start, flex-end, center, space-between, space-around */
.contenedor-horizontal {
    display: flex;
    justify-content: space-around; 
}

/* Ejercicio 2: Cambia align-items por: flex-start, flex-end, center */
.contenedor-vertical {
    display: flex;
    height: 250px; /* Altura para ver el movimiento vertical */
    align-items: center;
    justify-content: center; /* También centramos horizontal */
}

/* Ejercicio 3: Flex-wrap permite que los elementos bajen al faltar espacio */
.contenedor-wrap {
    display: flex;
    flex-wrap: wrap;
    gap: 15px; /* Espacio entre cajas */
}

```

### 👨‍🏫 Dinámica de Clase propuesta:

1. **El Juego de los Atributos:** en `style-2.css`, dentro de `.contenedor-horizontal`, cambien `space-around` por `space-between` y vean la diferencia con **Live Server**.
2. **El "Santo Grial" del Diseño:**Centrar algo perfectamente (vertical y horizontal) es tan fácil como poner:
* `display: flex;`
* `justify-content: center;`
* `align-items: center;`
